```python
import requests # pip install requests
from bs4 import BeautifulSoup as bs # pip install beautifulsoup4
import pandas as pd
import re
from datetime import datetime
import glob
```

```python
pd.options.display.width = 0
pd.set_option('display.width', 1000)
```

```python
# Load the webpage content

r = requests.get('https://live.euronext.com/en/markets/amsterdam')

# Convert to a Beautifulsoup object
soup = bs(r.content)

# Print out our html
contents = soup.prettify()
print(contents)
```

```python
info_box = soup.find(class_='isin_and_mic')
info_rows = info_box.find_all('td')
for row in info_rows:
    print(row.prettify())

```

```python
aex_info = {}

for index, row in enumerate(info_rows):
    if index == 0:
        aex_info['title'] = 'Euronext Amsterdam'
    elif index == 1:
        continue    
    else:
        content_key = 'Euronext Amsterdam'
        content_value = row.find(class_='nowrap').get_value()
        aex_info[content_key] = content_value
        
        
print(aex_info)
#print(len(info_rows))
```
